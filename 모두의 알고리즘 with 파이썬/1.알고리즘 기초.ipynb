#첫째 마당 알고리즘 기초

##1부터 N까지 연속한 숫자의 합을 구하는 알고리즘

```python
# 일반적인 방법

def sum_n(n):
    s = 0
    for i in range(1, n + 1):
        s = s + i
    return s
```

```python
# 가우스의 방법

def sum_n(n):
    return n * (n + 1) // 2
```

예를 들어 1부터 100까지의 합을 더해야 한다면,

1+100=101

2+99=101

3+98=101

...

50+51=101

-> 101이 50번 나오므로 101 x 50 = 5050

이를 일반화하면,
n(n+1)/2

###알고리즘 분석
첫번째 방법은 입력 크기 n에 대해 사칙연산(덧셈)을 n번 해야한다.

-> 필요한 계산 횟수가 입력 크기에 정비례할 때 O(n)이라 한다.

입력 크기 n에 따라 덧셈을 두 번씩 하는 알고리즘은 O(2n)으로 표기할 것 같지만, 이 알고리즘 역시 정비례하는 것이기 때문에 O(n)으로 표기한다.
O표기법은 정확한 계산 횟수를 산정하는 것이 아니라, 입력 크기와의 '관계'를 표현하는 것이기 때문이다.

두 번째 알고리즘은 입력 크기n과 무관하게 사칙 연산을 세 번만 하면 왼다. 이는 O(3)이지만, 앞서 말한 것처럼 관계를 나타내는 것이기 때문에 O(1)로 나타낸다.

O(1) : 필요한 계산 횟수가 입력 크기 n과 무관할 때

O(n) : 필요한 계산 횟수가 입력 크기 n과 비례할 때


##연습문제

```python
#1부터 n까지 연속한 숫자의 제곱의 합을 구하는 프로그램을 for문으로 구현하라.

def sum_sq(n):
  res_ = 0
  for i in range(1, n+1):
    res_ += i**2
  return res_

print(sum_sq(10))
```

위 문제의 계산복잡도는 무엇인가?

-> O(n)

##최댓값 찾기

```python
# List function

# insert(i,x) 리스트 i번 위치에 x를 추가
a = [1, 2, 3]
a.insert(0, 5)
print(a)


# pop(i) i번 위치에 있는 자료를 리스트에서 빼내면서 그 값을 함수의 결과값으로 돌려준다. i를 지정하지 않으면 맨 마지막 값을 빼서 돌려준다.
b = [1, 2, 3]
print(b.pop(0))
print(b.pop())


#x in a 어떤 자료 x가 리스트에 있는지 확인한다. <-> x not in a
c = [1, 2, 3]
print(1 in c)
print(5 in c)
print(5 not in c)
```

```python
# 최댓값 구하기

def find_max(a):
    n = len(a)    # 입력 크기 n
    max_v = a[0]  # 리스트의 첫 번째 값을 최댓값으로 기억
    for i in range(1, n):  # 1부터 n-1까지 반복
        if a[i] > max_v:   # 이번 값이 현재까지 기억된 최댓값보다 크면
            max_v = a[i]   # 최댓값을 변경
    return max_v

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max(v))
```

위 프로그램은 자료 n개 중에서 최댓값을 찾기 위해 비교를 n-1번 해야 한다.
즉 O(n-1) -> O(n)이 된다. (n이 굉장히 커지면 n번과 n-1번의 차이가 무의미해지므로)

```python
# 리스트에 숫자가 n개 있을 때 가장 큰 값이 있는 위치 번호를 돌려주는 알고리즘을 만들어라.

def find_max_idx(a):
    n = len(a)   # 입력 크기 n
    max_idx = 0  # 리스트 중 0번 위치를 일단 최댓값 위치로 기억
    for i in range(1, n):
        if a[i] > a[max_idx]:  # 이번 값이 현재까지 기억된 최댓값보다 크면
            max_idx = i        # 최댓값의 위치를 변경
    return max_idx

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_max_idx(v))
```

##연습문제

```python
# 숫자 n개를 리스트로 입력받아 최솟값을 구하는 프로그램을 만들어 보시오.

def find_min(a):
    n = len(a)
    min_ = a[0]
    for i in range(1, n):
        if a[i] < min_:
            min_ = a[i]
    return min_

v = [17, 92, 18, 33, 58, 7, 33, 42]
print(find_min(v))
```

##동명이인 찾기①

```python
# 자주 쓰는 집합 기능

#add(x) 집합에 자료 x를 추가한다.

#discard(x) 집합에 자료 X가 들어 있다면 삭제함.
s = {1, 2, 3}
s.discard(2)
print(s)

#x in s 어떤 자료 x가 s에 들어있는지 확인 (리스트와 동일)
```

```python
# 동명이인을 찾는 알고리즘

def find_same_name(a):
    n = len(a)   # 리스트의 자료 개수를 n에 저장
    result = set()  # 결과를 저장할 빈 집합
    for i in range(0, n - 1):      # 0부터 n-2까지 반복
        for j in range(i + 1, n):  # i+1부터 n-1까지 반복
            if a[i] == a[j]:       # 이름이 같으면
                result.add(a[i])   # 찾은 이름을 result에 추가
    return result

name = ["Tom", "Jerry", "Mike", "Tom"]
print(find_same_name(name))

name2 = ["Tom", "Jerry", "Mike", "Tom", "Mike"]
print(find_same_name(name2))
```

전체 비교 횟수 = 3+2+1+0=6

즉 1부터 n-1까지의 합이 전체 비교횟수가 되며 이를 1부터 n까지의 합을 구하는 공식에 n대신 n-1을 대입하면

$$\frac{1}{2}n^2-\frac{1}{2}n$$

이는 O(n^2)로 나타낸다.

##연습문제

```python
#n명중 두 명을 뽑아 짝을 짓는다고 할 때 모든 조합을 출력하는 알고리즘을 만들시오.

v = ["Tom", "Jerry", "Mike"]

n = len(v)

for i in range(0, n-1):
    for j in range(i+1, n):
        print(v[i]+'-'+v[j])
```

n명에서 두 명을 뽑아 짝으로 만들면 짝 조합이 

$\frac{n(n-1)}{2}$가지 출력된다. 이 경우의 수를 $nC_2$라고도 표현한다.
